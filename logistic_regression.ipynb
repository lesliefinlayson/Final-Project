```python
import pandas as pd
import os
```

```python
file_path=os.path.join('..', 'data', 'cleaned_data.csv')
model_df=pd.read_csv(file_path)  
```

```python
model_df.head()
```

```python
model_df.dtypes
```

```python
#set index
model_df = model_df.set_index('Job')
model_df.head()
```

```python
#drop unnecessary columns
model_df=model_df.drop(columns=['Customer', 'Start_Date', 'End_Date', 'Materials_Estimate', 'Materials_Actual_Cost', 'Labor_Estimate', 'Actual_Labor_Costs'])
model_df.head()
```

```python
#rename Gain? column
model_df.rename(columns = {"Gain?": "Gain"}, inplace = True)
model_df.head()
```

```python
#convert to format model can use
#convert object types to strings
model_df['Gain'] = model_df['Gain'].astype("string")
model_df['Location'] = model_df['Location'].astype("string")
model_df['Scope'] = model_df['Scope'].astype("string")
model_df.head()
```

```python
model_df.dtypes
```

```python
#convert to format model can use
#transform Gain column
def change_string(gain):
    if gain == "yes":
        return 1
    else:
        return 0
```

```python
model_df['Gain'] = model_df['Gain'].apply(change_string)
model_df.head()
```

```python
#transform Location column
def new_column(row):
    if row['Location'] == 'Aptos':
        return 1
    elif row['Location'] == 'Capitola':
        return 2
    elif row['Location'] == 'Felton':
        return 3
    elif row['Location'] == 'Los Gatos':
        return 4
    elif row['Location'] == 'Santa Cruz':
        return 5
    else:
        return 6
```

```python
model_df['Project_Location'] = model_df.apply(new_column, axis=1)
model_df
```

```python
#transform Scope column
def new_column(row):
    if row['Scope'] == 'bathroom':
        return 1
    elif row['Scope'] == 'deck':
        return 2
    elif row['Scope'] == 'fence':
        return 3
    elif row['Scope'] == 'garage':
        return 4
    elif row['Scope'] == 'kitchen':
        return 5
    elif row['Scope'] == 'landscaping':
        return 6
    else:
        return 7
```

```python
model_df['Project_Scope'] = model_df.apply(new_column, axis=1)
model_df.head()
```

```python
#drop duplicate columns
model_df=model_df.drop(columns=['Location', 'Scope'])
model_df.head()
```

```python
#change column order
model_df = model_df[['Project_Location', 'Project_Scope','Time_Months', 'Estimated_Total_Cost', 'Actual_Total_Cost', 'Gain_or_Loss', 'Gain']]
model_df.head()
```
