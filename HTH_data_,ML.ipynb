```python
#convert to format model can use
#convert object types to strings
final_df['Gain'] = final_df['Gain'].astype("string")
final_df['Location'] = final_df['Location'].astype("string")
final_df['Scope'] = final_df['Scope'].astype("string")
final_df.head()
```

```python
#convert to format model can use
#transform Gain column
def change_string(gain):
    if gain == "yes":
        return 1
    else:
        return 0
```

```python
final_df['Gain'] = final_df['Gain'].apply(change_string)
final_df.head()
```

```python
#transform Location column
def new_column(row):
    if row['Location'] == 'Aptos':
        return 1
    elif row['Location'] == 'Capitola':
        return 2
    elif row['Location'] == 'Felton':
        return 3
    elif row['Location'] == 'Los Gatos':
        return 4
    elif row['Location'] == 'Santa Cruz':
        return 5
    else:
        return 6
```

```python
final_df['Project_Location'] =final_df.apply(new_column, axis=1)
final_df.head()
```

```python
#transform Scope column
def new_column(row):
    if row['Scope'] == 'bathroom':
        return 1
    elif row['Scope'] == 'deck':
        return 2
    elif row['Scope'] == 'fence':
        return 3
    elif row['Scope'] == 'garage':
        return 4
    elif row['Scope'] == 'kitchen':
        return 5
    elif row['Scope'] == 'landscaping':
        return 6
    else:
        return 7
```

```python
final_df['Project_Scope'] = final_df.apply(new_column, axis=1)
final_df.head()
```

```python
#drop duplicate columns
final_df = final_df.drop(columns=['Location', 'Scope'])
final_df.head()
```

```python
#change column order
final_df = final_df[['Estimated_Total_Cost', 'Project_Scope','Time_Months', 'Project_Location', 'Gain']]
final_df.head()
```

```python
#rescale
final_df['Estimated_Total_Cost'] = final_df['Estimated_Total_Cost'] / 10000
final_df.head()
```

```python
compression_opts = dict(method='zip',
                        archive_name='out.csv')  
final_df.to_csv('cleaned_data.zip', index=True,
          compression=compression_opts)  
```

```python
#split the data into training and testing
# Define the features set.
X = final_df.copy()
X = X.drop("Gain", axis=1)
X.head()
```

```python
# Create target
y = final_df["Gain"]
y.head()
```

```python
X.describe()
```

```python
# Check the balance of our target values
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
   y, random_state=1, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs',
   max_iter=200,
   random_state=1)
```

```python
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred))
```
