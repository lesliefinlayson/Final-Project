```python
def new_column(row):
    if row['fname'] == 'Arlan':
        return 30
    elif row['fname'] == 'Gerald':
        return 50
    elif row['fname'] == 'Phil':
        return 25
    elif row['fname'] == 'Kate':
        return 15
    elif row['fname'] == 'Savanna':
        return 17.5
    elif row['fname'] == 'Chris':
        return 17.5
    elif row['fname'] == 'Warren':
        return 15
    elif row['fname'] == 'Daniel':
        return 15
    elif row['fname'] == 'Dave':
        return 27.5
    elif row['fname'] == 'Conor':
        return 17.5
    elif row['fname'] == 'Randolf':
        return 17.5
    elif row['fname'] == 'Brandon':
        return 15
    elif row['fname'] == 'Tommy':
        return 20
    elif row['fname'] == 'Don':
        return 15
    elif row['fname'] == 'Tyler':
        return 15  
    elif row['fname'] == 'DeWayne':
        return 15      
    else:
        return 15
```
