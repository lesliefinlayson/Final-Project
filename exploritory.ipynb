```python
file_path=os.path.join('..', 'data', 'receipts.csv')
receipts_df=pd.read_csv(file_path)
receipts_df.info()
```

```python
#in order to sum, convert cost column from object to float 
receipts_df["Cost"] = receipts_df["Cost"].replace('[\$\,\.]',"",regex=True).astype(float)
```

```python
receipts_df.info()
```

```python
receipts_df.head()
```

```python
receipts_df["Cost"] = receipts_df["Cost"].div(100).round(2)
```

```python
file_path=os.path.join('..', 'data', 'TSheets_data.csv')
TSheets_data_df=pd.read_csv(file_path) 
TSheets_data_df.info()
```

```python
receipts_df.head()
```

```python
#Find costs per job
##drop unneeded columns
material_costs_df=receipts_df.drop(columns=['Vendor', 'Description', 'Date', 'Unnamed: 5'])
material_costs_df
```

```python
#find costs per job
material_costs_df=material_costs_df.groupby(['Job']).sum()
material_costs_df.head()
```

```python
#look at bids spreadsheet and set index
Bids_df = Bids_df.set_index('Job')
Bids_df.head()
```

```python
#drop unnecessary columns for material estimate
materials_estimate_df=Bids_df.drop(columns=['Labor_Estimate', 'subtotal', 'overhead (15%)', 'total'])
materials_estimate_df.head()
```

```python
#merge materials cost df and materials estimate df
materials_df = pd.merge(material_costs_df, materials_estimate_df, left_index=True, right_index=True)
materials_df
```

```python
#change column order
materials_df = materials_df[['Location', 'Customer', 'Start_Date', 'End_Date', 'Scope', 'Materials_Estimate', 'Cost']]
```

```python
materials_df.head()
```

```python
#Rename column
materials_df.rename(columns = {"Cost": "Materials_Actual_Cost"}, inplace = True)
```
